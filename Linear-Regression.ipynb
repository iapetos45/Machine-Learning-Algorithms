```python
import os
import tarfile
import urllib

DOWNLOAD_ROOT = "https://raw.githubusercontent.com/rickiepark/handson-ml2/master/"
HOUSING_PATH = os.path.join("datasets", "housing")
HOUSING_URL = DOWNLOAD_ROOT + "datasets/housing/housing.taz"

def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
    taz_path = os.path.join(housing_url, taz_path)
    urllib.request.urlretrieve(housing_url, taz_path)
    housing_taz = tarfile.open(taz_path)
    housing_taz.extractall(path=housing_path)
    housing_taz.close()
    
    
fetch_hosing_data()
```
